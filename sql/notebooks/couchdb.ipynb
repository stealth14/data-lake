```python
'''========couchdb'=========='''
import couchdb 

#connection
user = 'admin'
password = 'naty'

serverUrl = 'http://{}:{}@127.0.0.1:5984/'.format(user,password);
server = couchdb.Server(serverUrl)

dbname = "covid"

if dbname in server:
    db = server[dbname]
else:
    db = server.create(dbname)
#read posgres
'''========postgresql'=========='''
import psycopg2
import psycopg2.extras

conn = psycopg2.connect("dbname=data user=postgres password=naty")
cur = conn.cursor(cursor_factory=psycopg2.extras.RealDictCursor)

def format_record(result):
    return {
        "id" : result["id"],
        "extended_tweet": result["extended_tweet"] if "extended_tweet" in result else None,
        "favorite_count": result["favorite_count"] if "favorite_count" in result else None,
        "retweet_count": result["retweet_count"] if "retweet_count" in result else None,
        "user_mentions": result["user_mentions"] if "user_mentions" in result else None,
    }

def postgres_tweets():
    try:
        cur.execute("SELECT * FROM data.tweets;")  
        for rec in cur:
          print(rec)
          docid,docrev = db.save(format_record(rec))
          print("saved : "+docid)
        
    except Exception as e:
        print(e)

postgres_tweets()

'''========mysql'=========='''

import pymysql.cursors

connection = pymysql.connect(host='localhost',
                             user='root',
                             db='data',
                             charset='utf8mb4',
                             cursorclass=pymysql.cursors.DictCursor)

def mysql_tweets():
    try:
        with connection.cursor() as cursor:
            cursor.execute("SELECT *  FROM `tweets`")
            for result in cursor:
                docid,docrev = db.save(result)
                print("saved : "+docid)
                
    except Exception as e:
        print(e)

mysql_tweets()
```
